# ðŸ“Š Trader Behavior vs Market Sentiment Analysis

This notebook analyzes trader behavior in relation to the Bitcoin Fear & Greed Index.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid', palette='muted')
```

## 1. Load Datasets

```python
# Load data
trader_df = pd.read_csv('/content/historical_data.csv')
sentiment_df = pd.read_csv('/content/fear_greed_index.csv')

# Show first few rows
print("Trader Data Sample:")
print(trader_df.head())

print("\nSentiment Data Sample:")
print(sentiment_df.head())
```

## 2. Clean & Preprocess Datasets

```python
# # No need to lowercase, but good practice to strip any hidden whitespace
# sentiment_df.columns = sentiment_df.columns.str.strip()

# # Select and rename columns
# sentiment_df = sentiment_df[['date', 'classification']].copy()
sentiment_df.rename(columns={'classification': 'sentiment'}, inplace=True)

# Convert to datetime
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])

# Optional: Map sentiment to binary
sentiment_df['sentiment_bin'] = sentiment_df['sentiment'].map({'Fear': 0, 'Greed': 1})


# --- Trader Data ---
trader_df['Timestamp IST'] = pd.to_datetime(trader_df['Timestamp IST'], dayfirst=True)
trader_df['date'] = pd.to_datetime(trader_df['Timestamp IST'].dt.date)

# Display info
print(trader_df.info())
print(sentiment_df.info())
```

## 3. Merge Datasets

```python
# Merge datasets on date
merged_df = pd.merge(trader_df, sentiment_df, on='date', how='left')

# Check for missing sentiments
print("Missing sentiment values:", merged_df['sentiment'].isnull().sum())

# Drop missing sentiment rows
merged_df.dropna(subset=['sentiment'], inplace=True)

# Preview merged dataset
merged_df.head()
```

## 4. Exploratory Data Analysis (EDA)

```python
# PnL by Sentiment
pnl_by_sentiment = merged_df.groupby('sentiment')['Closed PnL'].mean()
print("Average PnL by Sentiment:")
print(pnl_by_sentiment)

# Boxplot of PnL
plt.figure(figsize=(8,5))
sns.boxplot(x='sentiment', y='Closed PnL', data=merged_df)
plt.title("Trader PnL by Market Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Closed PnL")
plt.show()
```

```python
# Trade Side (Long/Short) vs Sentiment
if 'Side' in merged_df.columns:
    sns.countplot(x='Side', hue='sentiment', data=merged_df)
    plt.title("Trade Side by Market Sentiment")
    plt.show()
else:
    print("No 'Side' column found in trader data.")
```

```python
# Volatility clustering â€” high absolute PnL persistence
daily_vol = merged_df.groupby('date')['Closed PnL'].mean().abs().rename('volatility')
daily_vol_roll = daily_vol.rolling(window=7).mean()

plt.figure(figsize=(12,5))
daily_vol_roll.plot(label='7d rolling vol')
daily_vol.plot(alpha=0.3, label='daily vol')
plt.legend()
plt.title('Volatility Clustering (Absolute PnL)')
plt.show()
```

```python
#Statistical Testing: Fear vs Greed PnLs
from scipy.stats import ttest_ind

fear = merged_df[merged_df['sentiment']=='Fear']['Closed PnL']
greed = merged_df[merged_df['sentiment']=='Greed']['Closed PnL']

t_stat, p_value = ttest_ind(fear, greed, equal_var=False)
print(f"T-statistic: {t_stat:.3f}, pâ€‘value: {p_value:.3f}")
```

```python
#Enhanced Feature Engineering: Lagged & Rolling Metrics
daily_meta = merged_df.groupby('date')['Closed PnL'].mean().rename('pnl_mean').to_frame()
daily_meta['pnl_lag1'] = daily_meta['pnl_mean'].shift(1)
daily_meta['pnl_rollstd_7'] = daily_meta['pnl_mean'].rolling(7).std()
daily_meta['trade_count'] = merged_df.groupby('date')['Closed PnL'].count()
daily_meta['sentiment'] = sentiment_df.set_index('date')['sentiment']

daily_meta.dropna(inplace=True)
daily_meta.head()
```

```python
# Intraday Sentiment Patterns
merged_df['hour'] = trader_df['Timestamp IST'].dt.hour
hourly = merged_df.groupby(['hour','sentiment']).size().unstack(fill_value=0)
hourly_norm = hourly.div(hourly.sum(axis=1), axis=0)

hourly_norm.plot(kind='bar', stacked=True, figsize=(12,5))
plt.title('Hourly Trade Proportion by Sentiment')
plt.xlabel('Hour of Day')
plt.ylabel('Proportion')
plt.show()
```

```python
# Time-series aggregation with only two fields: pnl_mean and trade_count
daily_ts = merged_df.groupby('date').agg(
    pnl_mean = ('Closed PnL', 'mean'),
    trade_count = ('Closed PnL', 'count')
)

# Add rolling statistics
daily_ts['pnl_roll_mean'] = daily_ts['pnl_mean'].rolling(7).mean()
daily_ts['pnl_roll_std'] = daily_ts['pnl_mean'].rolling(7).std()

# Plot daily vs rolling average
plt.figure(figsize=(12,5))
daily_ts[['pnl_mean','pnl_roll_mean']].plot(title='Daily PnL & 7-day Rolling Mean')
plt.show()

# Plot trade count trend
daily_ts['trade_count'].plot(title='Daily Trade Count', figsize=(12,4))
plt.show()
```

```python
# Prepare data
corr_df = merged_df.copy()
corr_df['sentiment_bin'] = corr_df['sentiment'].map({'Fear': 0, 'Greed': 1})
corr_df['side_bin'] = corr_df['Side'].map({'Long': 1, 'Short': 0})

# Compute correlations
static_corr = corr_df[['Closed PnL','side_bin','sentiment_bin']].corr()
sns.heatmap(static_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix: PnL, Side, Sentiment')
plt.show()

```

```python
# Ensure columns exist and are standardized
print(corr_df.columns)
assert 'Closed PnL' in corr_df.columns
assert 'sentiment' in corr_df.columns

# Filter to valid numeric PnL rows and non-null sentiment
plot_df = corr_df.loc[
    corr_df['Closed PnL'].notnull() & corr_df['sentiment'].notnull(),
    ['Closed PnL', 'sentiment']
].copy()

# Make sure sentiment is categorical
plot_df['sentiment'] = plot_df['sentiment'].astype(str)

# Plot KDE distributions
plt.figure(figsize=(10,6))
sns.kdeplot(
    data=plot_df,
    x='Closed PnL',
    hue='sentiment',
    common_norm=False,
    fill=True,
    alpha=0.5
)
plt.title('PnL Distribution by Market Sentiment')
plt.xlabel('Closed PnL')
plt.ylabel('Density')
plt.show()
```

 # Model Evaluation & Visuals

```python


from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_curve, auc, precision_recall_curve, roc_auc_score
)

# --- 1. Compute daily aggregates (without leverage)
daily = merged_df.groupby('date').agg({
    'Closed PnL': ['mean', 'std', 'count'],
    'Side': lambda x: (x == 'Long').mean()
})
daily.columns = ['pnl_mean', 'pnl_std', 'trade_count', 'long_ratio']
daily = daily.reset_index()

# --- 2. Attach sentiment labels
daily = daily.merge(
    sentiment_df[['date', 'sentiment']], on='date', how='left'
).dropna(subset=['sentiment'])

# --- 3. Cluster the behavior of days
features = daily[['pnl_mean', 'long_ratio']]
kmeans = KMeans(n_clusters=3, random_state=42).fit(features)
daily['cluster'] = kmeans.labels_

# --- 4. Prepare data for classification
X = features
y = (daily['sentiment'] == 'Greed').astype(int)

# === Here's where the split happens ===
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- 5. Train the model
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# --- 6. Model evaluation
y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:, 1]

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Pred: Fear', 'Pred: Greed'],
            yticklabels=['Actual: Fear', 'Actual: Greed'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}', color='darkorange')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()

# Precisionâ€“Recall curve
precision, recall, _ = precision_recall_curve(y_test, y_prob)
plt.figure(figsize=(6,6))
plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precisionâ€“Recall Curve')
plt.show()

# Final performance metrics
print("=== Classification Report ===")
print(classification_report(y_test, y_pred))
print("ROCâ€‘AUC score:", round(roc_auc_score(y_test, y_prob), 3))

# Feature Importances
importances = pd.Series(clf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nFeature importances:\n", importances)

# Clusterâ€“sentiment breakdown
print("\nCluster by Sentiment Distribution:")
print(daily.groupby(['cluster', 'sentiment']).size().unstack(fill_value=0))
```
